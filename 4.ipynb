```python
# Открываем изображение
img = cv2.imread('cat_text.jpg', 1)

mask1 = np.zeros(img.shape[:2], np.uint8)
mask1[100:300, 100:400] = 255
masked_img1 = cv2.bitwise_and(img, img, mask = mask1)


color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([masked_img1],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```

```python
img = cv2.imread('dog_3.jpeg')
# equalize the histogram of the input image

img_yuv = cv2.cvtColor(img, cv2.COLOR_BGR2YUV)
# equalize the histogram of the Y channel
img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
# convert the YUV image back to RGB format
img_output = cv2.cvtColor(img_yuv, cv2.COLOR_YUV2BGR)


cv2.imshow("Original", img)
cv2.imshow("Image", img_output)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img = cv2.imread('cat_text.jpg', 0)

# Создаем список допустимых значений параметра type метода cv2.threshold
types = [cv2.THRESH_BINARY, cv2.THRESH_BINARY_INV, cv2.THRESH_TRUNC, cv2.THRESH_TOZERO, cv2.THRESH_TOZERO_INV]

# Пройдемся по всем допустимым значениям type
for type in types:
    # Создаем копию исходного изображения
    img_copy = img.copy()

    # Применяем бинаризацию с текущим значением type
    ret, thresh = cv2.threshold(img_copy, 127, 255, type)

    cv2.imshow("Image", thresh)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

№5

```python
# Открываем изображение
img = cv2.imread('cat_text.jpg', 0)

# Создаем список различных комбинаций значений параметров maxValue, blockSize, thresholdType метода adaptiveThreshold
maxValues = [20, 150] #интенсивность цвета для пикселей, которые превысили пороговое значение
blockSizes = [9, 11] #Размер области соседних пикселей, участвующих в нахождении пороговог значения
thresholdTypes = [cv2.ADAPTIVE_THRESH_MEAN_C, cv2.ADAPTIVE_THRESH_GAUSSIAN_C]

# Пройдемся по всем комбинациям
for maxValue in maxValues:
    for blockSize in blockSizes:
        for thresholdType in thresholdTypes:
            # Создаем копию исходного изображения
            img_copy = img.copy()

            # Применяем адаптивную бинаризацию с текущими значениями maxValue, blockSize, thresholdType
            thresh = cv2.adaptiveThreshold(img_copy, maxValue, thresholdType, cv2.THRESH_BINARY, blockSize, 2)

            cv2.imshow("Image", thresh)
            cv2.waitKey(0)
            cv2.destroyAllWindows()
```
